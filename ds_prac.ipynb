As there are some null values in the dataset we will fill the null values with function fillna

As we have couple of null values in categorical data we will use the direct filling method.

```python
data['Name'].fillna("FIFA Soccer 07",inplace=True)
```

```python
data['Genre'].fillna("Sports",inplace=True)
```

```python
data['Publisher'].fillna("Nintendo",inplace=True)
```

We have null values in numerical data so we will use mean method to fill the null values

```python
mean=int(data["Year_of_Release"].mean())
data["Year_of_Release"].fillna(mean,inplace=True)
```

Now all the null values are filled

```python
#Pie plot for genre
plt.figure(figsize=(10,12))
data['Genre'].value_counts().plot.pie(autopct='dark')
plt.show()
```

Genre Diversity: The presence of numerous slices indicates a diversified landscape of video game genres within the dataset. This variety caters to a broad spectrum of player preferences and interests.
Predominant Genres: Based on slice size, "Action" and "Adventure" emerge as the most dominant genres, collectively capturing a significant portion of the total games. This dominance might reflect their widespread appeal or high commercial viability.
Proportionately Smaller Genres: Genres like "Simulation", "Strategy", and "Puzzle" occupy smaller slices, suggesting a niche audience or lower production volume compared to the leaders. However, their presence highlights the existence of catering to specialized gaming tastes.

```python
fig,ax=plt.subplots(figsize=(15,8))
ax.scatter(data['Year_of_Release'],data['EU_Sales'])
ax.set_xlabel('Year_of_Release')
ax.set_ylabel('EU_Sales')
plt.show()
```

Release Trends: The lines show how the number of games released in each genre changes over time. Some genres might exhibit increasing or decreasing trends in releases, while others might remain relatively stable.
Genre Comparison: By comparing the lines, you can see how the release volume of different genres compares to each other across different years. Some genres might consistently have more releases than others, or their popularity might fluctuate over time.

```python
sns.lineplot(x='Year_of_Release',y='NA_Sales',label='NA_Sales',data=data)
plt.legend()
plt.show()
```

Sales Trends: The lines show how the average sales of games change over time. Some genres or
platforms might exhibit increasing or decreasing trends in sales, while others might remain
relatively stable.
Genre/Platform Comparison: By comparing the lines, you can see how the average sales of different

```python
data.groupby('Platform')[['JP_Sales','EU_Sales','NA_Sales']].sum().plot.bar(color=['green','blue','pink'],figsize=(20,5))
plt.ylabel('count')
plt.show()
```

Global Game Release Landscape: The plot reveals a diverse landscape of countries where video games are released, indicating a global industry with a widespread reach.
Predominant Releasing Countries: Some countries appear to have released significantly more games than others, based on the bar heights. However, due to the image resolution, it's difficult to identify the specific countries and their exact release counts.
Proportionately Smaller Releasing Countries: Some countries have fewer games released, suggesting a smaller domestic game development industry or focus on different entertainment sectors.

```python
data.groupby('Genre')[['NA_Sales','JP_Sales']].sum().plot.bar(color=['green','red'],figsize=(8,5))
plt.ylabel('sales as per genre in JP and NA')
plt.show()
```

Action: The most dominant genre across almost all platforms, highlighting its enduring popularity among gamers.
Adventure: Another widely popular genre, with a strong presence on most platforms except "Atari 2600" and "NES".
RPG: Primarily concentrated on PC, "PlayStation", and "Xbox" platforms, suggesting a preference for more powerful hardware for these often resource-intensive games.
Sports: Particularly prominent on "Wii", "PS2", and "Xbox 360", potentially due to the motion control and multiplayer capabilities of these platforms being well-suited for sports games.
Strategy: Primarily found on PC, likely due to its traditional association with keyboard and mouse controls and complex gameplay mechanics often favored by strategy game players.
